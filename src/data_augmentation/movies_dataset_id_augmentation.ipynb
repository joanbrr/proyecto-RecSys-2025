```python
movies_df = pd.read_csv("../../data/interim/movies.csv", sep=',', header=0)
links_df = pd.read_csv("../../data/interim/links.csv", sep=',', header=0)
```

```python
i_cols = [
    'item_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL',
    'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy',
    'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror',
    'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'
]

items_df = pd.read_csv('../../data/raw/ml-100k/u.item', sep='|', names=i_cols, encoding='latin-1')

items_df = items_df.drop(['item_id', 'video_release_date'], axis=1)
```

Necesitamos juntar movies_df con items_df por el título y luego el dataframe resultante con el links_df por el movieId

```python
merged_df = movies_df.merge(items_df, on='title', how='inner')
```

```python
final_df = merged_df.merge(links_df, on='movieId', how='inner')
final_df = final_df.drop(['IMDb_URL', 'genres'], axis=1)
```
